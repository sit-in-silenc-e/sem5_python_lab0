```python
def task_3(input_data=None):
    if input_data is None:
        input_data = input("Введіть список чисел через пробіл: ")
    
    try:
        numbers = list(map(int, input_data.split()))
    except ValueError:
        return "Помилка: введіть коректні цілі числа!"
    
    if not numbers:
        return "Помилка: список не може бути порожнім!"
    
    numbers.sort()
    n = len(numbers)
    
    if n % 2 == 1:
        return numbers[n // 2]
    else:
        return (numbers[n // 2 - 1], numbers[n // 2])
```

```python
print("Тест 1: 2 4 5 1 7")
result = task_3("2 4 5 1 7")
print(f"Очікуваний результат: 4, Отриманий результат: {result}")
```

```python
print("Тест 2: 7 3 1 7 4 2 8 2")
result = task_3("7 3 1 7 4 2 8 2")
print(f"Очікуваний результат: (3, 4), Отриманий результат: {result}")
```

```python
print("Тест 3: некоректні дані")
result = task_3("2 4 a b")
print(f"Очікуваний результат: Помилка, Отриманий результат: {result}")
```
